```python
import pandas as pd
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['pakistan'])
# Interest by Region
df = pytrend.interest_by_region()
print(df.iloc['pakistan'])
```

```python
df.reset_index().plot(x='geoName', y='pakistan', figsize=(120, 10), kind ='bar')
```

```python
# Get Google Hot Trends data
df = pytrend.trending_searches(pn='Pk')
df.head()
```

```python
df = pytrend.today_searches(pn='Pk')
```

```python
# Get Google Top Charts
df = pytrend.top_charts(2022, hl='en-US', tz=300, geo='PK')
df = pytrend.top_charts(2020, hl='en-US', tz=300, geo='PK')
df
```
