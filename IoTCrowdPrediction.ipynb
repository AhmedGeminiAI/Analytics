```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# I have taken a hypothetical situation based over some probabilities
# There are three rooms in a building which are hosting an event. The visitors can go to any of these rooms while freely walking
# Data shows that
# 10% of Room1 visitorsrs will switch to Room2 and 10% to Room3.
# 20% of Room1 visitors will switch to Room2 and 20% to Room3.
# 10% of Room1 visitors will switch to Room2 and 20% to Room3 .
# After significantly long time we will be finding the  % of the visitors will stay in the same room.
#https://github.com/punitzen/markov-chain-application-using-python/tree/main
```

```python
# P is transition state matrix
P = np.array([[0.8,0.1,0.1],
              [0.2,0.6,0.2],
              [0.1,0.2,0.7]])
```

```python
state=np.array([[1.0,0.0,0.0]])
stateHist=state
dfStateHist=pd.DataFrame(state)
distr_hist = [[0,0,0]]

for x in range(100):
    state=np.dot(state,P)
    stateHist=np.append(stateHist,state,axis=0)
    dfDistrHist = pd.DataFrame(stateHist)
```

```python
# Plotting graph
plt.figure(figsize=(10,5))
print(dfDistrHist)
sns.lineplot(data=dfDistrHist)
```
